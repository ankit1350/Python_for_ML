* **aggregate** : reduces a set of values into a single summary value.
* It is used to summarize and analyze data
* Often used with the groupby() function

```python
import pandas as pd
```

```python
df=pd.read_csv("data.csv")
df
```

* find mean of any column that is numeric

```python
m=df.mean(numeric_only=True)
m
```

```python
s=df.sum(numeric_only=True)
s
```

```python
m1=df.min(numeric_only=True)
m1
```

```python
m2=df.max(numeric_only=True)
m2
```

```python
c=df.count()
c
```

* get aggregate value for a particular column

```python
cc=df["Speed"].mean()
cc
```

```python
df["Defense"].sum()
```

```python
group=df.groupby("Type1")
group
```

* get the mean according to the Type 1 

```python
group["Defense"].mean()
```

```python
group["Speed"].sum()
```

```python
group["Speed"].min()
```

```python
group["Speed"].count()
```
