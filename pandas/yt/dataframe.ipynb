### Dataframe is a 2D representation

```python
import pandas as pd
```

```python
data={"Name":['spongebob','patrick','squidward'],
      "Age":[30,35,49],
      }
```

```python
df=pd.DataFrame(data)
df
```

```python
df=pd.DataFrame(data, index=["employee1","employee2","employee3"])
```

```python
df
```

```python
df.loc["employee1"]
```

```python
df.iloc[0]
```

* adding a new column

```python
df["job"]=['cook','N/A','cashier']
df
```

* adding a new row

```python
new_row=pd.DataFrame([{"Name":"sandy", "Age":23,"job":"nothing"}],index=["employee4"])
```

```python
df=pd.concat([df,new_row])
```
