```python
student= {
    "names": [
        "Ankit",
        "Rohit",
        "Sneha",
        "Priya",
        "Aman"
    ],

    "marks": [
        [85, 78, 90],   # Ankit → Maths, Physics, CS
        [70, 88, 75],   # Rohit
        [92, 91, 89],   # Sneha
        [60, 65, 58],   # Priya
        [80, 85, 82]    # Aman
    ],

    "attendance_percent": [
        92.5,
        88.0,
        96.3,
        75.4,
        90.1
    ]
}
```

```python
s= pd.DataFrame(student)
```

```python
s
```

```python
s.groupby('attendance_percent').sum(numeric_only=True)
```

```python
s['total_marks'] = s['marks'].apply(sum)
```

```python
s
```

* describe gives all the mathematical statistics related to thee data

```python
s.describe().T
```

* **count** → number of non-NaN (valid) values
* **mean** → average of the values
* **std** → standard deviation, showing data spread around the mean
* **min** → smallest value in the column
* **25%** → first quartile (25% of values are below or equal to this)
* **50%** → median (middle value of the data)
* **75%** → third quartile (75% of values are below or equal to this)
* **max** → largest value in the column
